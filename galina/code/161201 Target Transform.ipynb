```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
train_Y = pd.read_csv('../data/train_Y.csv')
```

```python
plt.figure(figsize = (4, 3))
sns.distplot(train_Y)
```

```python
plt.figure(figsize = (4, 3))
sns.distplot(np.log(train_Y))
plt.xlim((4, 11))
```

```python
plt.figure(figsize = (4, 3))
sns.distplot(np.log(train_Y + 200))
plt.xlim((4, 11))
```

```python
plt.figure(figsize = (4, 3))
sns.distplot(np.log(train_Y ** 0.25))
plt.xlim((1, 3))
```

```python
train_Y_log_200 = np.log(train_Y + 200)
train_Y_log_200.to_csv('../data/train_Y_log_200.csv', index = False)
```
