```python
from __future__ import division

import pandas as pd
from sklearn import preprocessing
```

```python
sample_sub = pd.read_csv('../data/sample_submission.csv')
test       = pd.read_csv('../data/test.csv')
train      = pd.read_csv('../data/train.csv')
```

```python
train_X = train.drop('loss', axis = 1)
train_Y = train['loss']

test_X  = test.copy()

# Convert categorical string columns into integers. There are 116 categorical columns.
# In a few cases test_X categorical column contains values that are absent from 
# corresponding train_X column. In such cases I replace these values with the most
# common value in test_X column.

le = preprocessing.LabelEncoder()
for col in train_X:
    if train_X[col].dtype == 'object':
        print col
        le.fit(train_X[col])
        
        try:
            test_X[col]  = le.transform(test_X[col])
        except ValueError:
            replacement = test_X[col].value_counts().index[0]
            values = train_X[col].unique()
            test_X.ix[~test_X[col].isin(values), col] = replacement
            test_X[col]  = le.transform(test_X[col])

        train_X[col] = le.transform(train_X[col])
        
train_X.to_csv('../data/train_X.csv', index = False)
train_Y.to_csv('../data/train_Y.csv', index = False, header = True)
test_X.to_csv('..data/test_X.csv', index = False)
```
